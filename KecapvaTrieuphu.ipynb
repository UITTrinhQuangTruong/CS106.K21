```
import math

class TrangThai(): #Trang Thai
	def __init__(self, KeCuopLeft, TrieuPhuLeft, Thuyen, KeCuopRight, TrieuPhuRight):
		self.KeCuopLeft = KeCuopLeft
		self.TrieuPhuLeft = TrieuPhuLeft
		self.Thuyen = Thuyen
		self.KeCuopRight = KeCuopRight
		self.TrieuPhuRight = TrieuPhuRight
		self.Cha = None

	def HoanThanh(self):
		if self.KeCuopLeft == 0 and self.TrieuPhuLeft == 0:
			return True
		else:
			return False

	def DKRangBuoc(self):
		if self.TrieuPhuLeft >= 0 and self.TrieuPhuRight >= 0 \
                   and self.KeCuopLeft >= 0 and self.KeCuopRight >= 0 \
                   and (self.TrieuPhuLeft == 0 or self.TrieuPhuLeft >= self.KeCuopLeft) \
                   and (self.TrieuPhuRight == 0 or self.TrieuPhuRight >= self.KeCuopRight): #So Trieu phu > so Ke cuop
			return True
		else:
			return False

	def __eq__(self, other): #So sanh 2 trang thai
		return self.KeCuopLeft == other.KeCuopLeft and self.TrieuPhuLeft == other.TrieuPhuLeft \
                   and self.Thuyen == other.Thuyen and self.KeCuopRight == other.KeCuopRight \
                   and self.TrieuPhuRight == other.TrieuPhuRight

	def __hash__(self):
		return hash((self.KeCuopLeft, self.TrieuPhuLeft, self.Thuyen, self.KeCuopRight, self.TrieuPhuRight))

def Next(TT_cu): # Xet trang thai ke tiep
	Con = [];
	if TT_cu.Thuyen == 'left': 
		TT_moi = TrangThai(TT_cu.KeCuopLeft, TT_cu.TrieuPhuLeft - 2, 'right',
                                  TT_cu.KeCuopRight, TT_cu.TrieuPhuRight + 2)
		# Di chuyen 2 Trieu Phu
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft - 2, TT_cu.TrieuPhuLeft, 'right',
                                  TT_cu.KeCuopRight + 2, TT_cu.TrieuPhuRight)
		## Two cannibals cross left to right.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft - 1, TT_cu.TrieuPhuLeft - 1, 'right',
                                  TT_cu.KeCuopRight + 1, TT_cu.TrieuPhuRight + 1)
		## One missionary and one cannibal cross left to right.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft, TT_cu.TrieuPhuLeft - 1, 'right',
                                  TT_cu.KeCuopRight, TT_cu.TrieuPhuRight + 1)
		## One missionary crosses left to right.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft - 1, TT_cu.TrieuPhuLeft, 'right',
                                  TT_cu.KeCuopRight + 1, TT_cu.TrieuPhuRight)
		## One cannibal crosses left to right.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
	else:
		TT_moi = TrangThai(TT_cu.KeCuopLeft, TT_cu.TrieuPhuLeft + 2, 'left',
                                  TT_cu.KeCuopRight, TT_cu.TrieuPhuRight - 2)
		## Two missionaries cross right to left.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft + 2, TT_cu.TrieuPhuLeft, 'left',
                                  TT_cu.KeCuopRight - 2, TT_cu.TrieuPhuRight)
		## Two cannibals cross right to left.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft + 1, TT_cu.TrieuPhuLeft + 1, 'left',
                                  TT_cu.KeCuopRight - 1, TT_cu.TrieuPhuRight - 1)
		## One missionary and one cannibal cross right to left.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft, TT_cu.TrieuPhuLeft + 1, 'left',
                                  TT_cu.KeCuopRight, TT_cu.TrieuPhuRight - 1)
		## One missionary crosses right to left.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
		TT_moi = TrangThai(TT_cu.KeCuopLeft + 1, TT_cu.TrieuPhuLeft, 'left',
                                  TT_cu.KeCuopRight - 1, TT_cu.TrieuPhuRight)
		## One cannibal crosses right to left.
		if TT_moi.DKRangBuoc():
			TT_moi.Cha = TT_cu
			Con.append(TT_moi)
	return Con

def BFS():
	TT_bandau = TrangThai(3,3,'left',0,0)
	if TT_bandau.HoanThanh():
		return TT_bandau
	queueTT = list()
	flagsTT = set()
	queueTT.append(TT_bandau) # FIFO vao truoc ra truoc
	while queueTT: # KHi no chua rong
		tempTT = queueTT.pop(0)
		if tempTT.HoanThanh():
			return tempTT
		flagsTT.add(tempTT)
		Con = Next(tempTT)
		for i in Con:
			if (i not in flagsTT) or (i not in queueTT):
				queueTT.append(i)
	return None

def InKQ(KQ):
		path = []
		path.append(KQ)
		Cha = KQ.Cha
		while Cha:
			path.append(Cha)
			Cha = Cha.Cha

		for t in range(len(path)):
			tempTT = path[len(path) - t - 1]
			print ("(" + str(tempTT.KeCuopLeft) + "," + str(tempTT.TrieuPhuLeft) + "," + tempTT.Thuyen + "," + str(tempTT.KeCuopRight) + "," + str(tempTT.TrieuPhuRight) + ")")

KQ = BFS()
print ("TRIEU PHU VA KE CUOP:")
print ("(KeCuopLeft,TrieuPhuLeft,Thuyen,KeCuopRight,TrieuPhuRight)")
InKQ(KQ)
```
